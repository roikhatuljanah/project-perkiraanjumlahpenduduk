```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_surabaya = pd.read_csv('kotasurabaya.csv')
```

```python
df_surabaya
```

```python
df_surabaya.describe()
```

```python
df_surabaya.info()
```

```python
plt.scatter(df_surabaya['Tahun'], df_surabaya['Jumlah'])
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
```

```python
x = df_surabaya.iloc[:,[0]].values
y = df_surabaya.iloc[:,[1]].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regresi = LinearRegression()
```

```python
regresi.fit(x_train,y_train)
```

```python
y_prediksi = regresi.predict(x_test)
```

```python
y_prediksi
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
print("MAE : ",mean_absolute_error(y_test,y_prediksi))
print("R2 : ",r2_score(y_test,y_prediksi))
```

```python
data_baru = pd.read_csv('datatestprediksi.csv')
```

```python
prediksi_baru = regresi.predict(data_baru)
prediksi_baru
```

```python
plt.figure(figsize=(15,10))
plt.title("Visualisasi Data Hasil Prediksi dari data training")

plt.scatter(x_train,y_train,c='blue')
plt.plot(x_train,regresi.predict(x_train),c='red')
plt.xlabel("Tahun")
plt.ylabel("Jumlah")
plt.grid()
```

```python
plt.figure(figsize=(15,10))
plt.title("Visualisasi Data Hasil Prediksi dari data training")

plt.scatter(x_test,y_test,c='green')
plt.plot(x_test,y_prediksi,c='yellow')

plt.scatter(data_baru, prediksi_baru, c='red')
plt.xlabel("Tahun")
plt.ylabel("Jumlah")
plt.grid()
```
